```python
import numpy as np
from numpy import *
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.patheffects as pe
pek=[pe.Stroke(linewidth=5, foreground='k',alpha=1), pe.Normal()]
from matplotlib import colors

alph = 0.5
col1 = 'firebrick'
col2 = 'seagreen'
col3 = 'navy'
col1a = array(colors.to_rgba(col1))
col1a[-1] = alph
col2a = array(colors.to_rgba(col2))
col2a[-1] = alph
col3a =array(colors.to_rgba(col3))
col3a[-1] = alph


fig = plt.figure(figsize=(10,10))
ax = fig.gca(projection='3d')


# create x,y
xx, zz = np.meshgrid(linspace(-1,1,2),linspace(-1,1,2))
yy = zeros(shape=(2,2))
ax.plot_surface(xx, yy, zz,edgecolor=col2,color=col2a,lw=5)


xx, yy = np.meshgrid(linspace(-1,1,2),linspace(-1,1,2))
zz = zeros(shape=(2,2))
ax.plot_surface(xx, yy, zz,edgecolor=col1,color=col1a,lw=5)

yy, zz = np.meshgrid(linspace(-1,1,2),linspace(-1,1,2))
xx = zeros(shape=(2,2))
ax.plot_surface(xx, yy, zz,edgecolor=col3,color=col3a,lw=5)

ax.elev = 30
ax.azim = 10

plt.axis('off')

plt.savefig('2Daxes.pdf', transparent=True)
```
